# Generate map

```python
import random
import numpy as np

# generate random points on map with given deviation
lat, long = (49.87087, 11.08979)
dev = 1 / 500
data = np.array([[lat + (dev + dev) * i / 30 - dev, long + random.uniform(-dev, dev), - random.randint(0, 100)] for i in range(30)])

x_map.plot(data)
```
